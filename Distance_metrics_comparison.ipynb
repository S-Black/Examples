```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

import hdbscan
from sklearn.cluster import KMeans
```

```python
iris = px.data.iris()
df = iris.drop(["species", "species_id"], axis=1)
```

```python
print('Actual data')
plt.scatter(df.sepal_length, df.sepal_width, c=iris.species_id, cmap="gist_rainbow")
plt.xlabel("Spea1 Length", fontsize=18)
plt.ylabel("Sepal Width", fontsize=18)
plt.show()

for m in hdbscan.dist_metrics.METRIC_MAPPING:
    if m not in [
        "minkowski",
        "seuclidean",
        "mahalanobis",
        "wminkowski",
        "haversine",
        "cosine",
        "arccos",
        "pyfunc",
    ]:
        clusterer = hdbscan.HDBSCAN(min_cluster_size=2, metric=m)
        clusterer.fit(df)

        print(
            "Distance metric: ",
            m,
            ", Number of clusters: ",
            len(np.unique(clusterer.labels_)),
        )

        plt.scatter(
            df.sepal_length, df.sepal_width, c=clusterer.labels_, cmap="gist_rainbow"
        )
        plt.xlabel("Spea1 Length", fontsize=18)
        plt.ylabel("Sepal Width", fontsize=18)
        plt.show()
```

```python
n_clusters = 3

algorithm = KMeans
model = algorithm(n_clusters).fit(df)

plt.scatter(df.sepal_length, df.sepal_width, c=model.labels_, cmap="gist_rainbow")
plt.xlabel("Spea1 Length", fontsize=18)
plt.ylabel("Sepal Width", fontsize=18)
plt.show()
```
