```python
#cumulative
fig = go.Figure()

'''
for year in range(2015,2020):
    fig.add_trace(go.Scatter(x=df['Date'], y=df[year],
                        #name=str(year),
                        showlegend=False,
                        line=dict(color='lightgrey')))
'''
fig.add_trace(go.Scatter(x=df['Date'], y=df['max'].cumsum(),
                         name='maximum & minimum',
                        line=dict(color='grey', dash='dot')))
fig.add_trace(go.Scatter(x=df['Date'], y=df['min'].cumsum(),
                         showlegend=False,
                        line=dict(color='grey', dash='dot')))
fig.add_trace(go.Scatter(x=df['Date'], y=df['5yr_avg'].cumsum(),
                         name='5 year average',
                        line=dict(color='grey')))

fig.add_trace(go.Scatter(x=df['Date'], y=df[2020].cumsum(),
                         name='2020',
                        line=dict(color='crimson')))
    
fig.update_layout(xaxis_tickformat='%b',
                 template='simple_white',
                 yaxis_type='log')

fig.show()
```
